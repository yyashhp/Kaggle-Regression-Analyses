```python
import pandas as pd, numpy as np, statsmodels.formula.api as stats, statistics 
```

```python
housing_dataframe = pd.read_csv('NY-House-Dataset.csv')
```

```python
housing_dataframe
```

## We will attempt to use these some columns to predict housing price. First, I want to collect some of the relevant variables into a seperate dataframe

```python
housing_dataframe.columns.values #to get a list of the column titles
```

```python
X = housing_dataframe[['TYPE', 'BEDS', 'BATH', 'PROPERTYSQFT', 'SUBLOCALITY']]
Y = housing_dataframe['PRICE']
```

```python
def convert_type(val):
    return str(val).split()[0]
X['TYPE'] = X['TYPE'].apply(lambda x: convert_type(x))
X['BATH'] = X['BATH'].apply(lambda x: round(x, 3))
X['PROPERTYSQFT']  = X['PROPERTYSQFT'].apply(lambda x: round(x, 3))
print(len(X))
X.dropna(axis = 1, inplace = True)
print(len(X))
```

```python
Y.dropna(inplace = True)
```

```python
data = pd.concat((Y,X), axis = 1)
data
```

```python
model = stats.ols(formula='PRICE ~ TYPE + BEDS + BATH + PROPERTYSQFT + SUBLOCALITY' , data = data).fit()
model.predict(X)
```

```python
dir(model)
```

```python
model.summary()
```

```python
data_2 = data['SUBLOCALITY'].apply(lambda x: x if (x == "New York County" or x == "Richmond County") else "Other")
data_2 = pd.concat((data_2, data.drop(columns = 'SUBLOCALITY')), axis = 1)
data_2
```

```python
model_2 = stats.ols(formula = 'PRICE ~ SUBLOCALITY + TYPE + BEDS + BATH + PROPERTYSQFT', data = data_2).fit()
model_2.summary()
```

```python
data_3 = data_2.drop(columns = ['SUBLOCALITY', 'TYPE'])
model_3 = stats.ols(formula = 'PRICE ~ BEDS + BATH + PROPERTYSQFT', data = data_3).fit()
model_3.summary()
```

```python
bath_data = data_3['BATH'].tolist()
bed_data = data_3['BEDS'].tolist()
np.corrcoef(bath_data,bed_data)
# We will merge to bath+bed
data_3['B_B'] = data_3['BATH'] + data_3['BEDS']
data_4 = data_3.drop(columns = ['BATH', 'BEDS'])
model_4 = stats.ols(formula = 'PRICE ~ B_B + PROPERTYSQFT', data = data_4).fit()
#model_4.summary()
model_5 = stats.ols(formula = 'np.log(PRICE) ~ B_B +  np.log(PROPERTYSQFT)', data = data_4).fit()
model_5.summary()
```

```python
print(data_2.loc[data_2['TYPE'] == "Foreclosure"].index.tolist())

data_2.drop(labels = data_2.loc[data_2['TYPE'] == "Foreclosure"].index.tolist(), axis = 0, inplace = True)
```

```python
data_2
```

```python
model_6 = stats.ols(formula = 'np.log(PRICE) ~ TYPE + B_B + np.log(PROPERTYSQFT)', data = pd.concat((data_4, data_2.drop(columns = ['SUBLOCALITY', 'BEDS', 'BATH', 'PRICE'])), axis = 1)).fit()
model_6.summary()
```

## At this point, R^2 is okay but not that good. Let's do some diagnostics on these variables and see if we notice anything

```python
curr_model = pd.concat((data_4, data_2.drop(columns = ['SUBLOCALITY', 'BEDS', 'BATH', 'PRICE', 'PROPERTYSQFT'])), axis = 1)
#curr_model['PROPERTYSQFT'] = np.log(curr_model['PROPERTYSQFT'])
#curr_model['PRICE'] = np.log(curr_model['PRICE'])
curr_model.drop(columns = 'B&B', inplace = True)
curr_model
```

```python
np.corrcoef([curr_model['PROPERTYSQFT'],curr_model['B_B']])
pp_bb_test = curr_model
pp_bb_test['mult'] = np.log(pp_bb_test['PROPERTYSQFT']) * pp_bb_test['B_B']
pp_bb = stats.ols('mult ~ PROPERTYSQFT + B_B', data = pp_bb_test).fit()
pp_bb.summary()
model_7 = stats.ols(formula = 'np.log(PRICE) ~ TYPE + mult + np.log(PROPERTYSQFT)', data = pp_bb_test).fit()
model_7.summary()
```

## Lets try adding back some other features now that we have a good base


```python
newest_data = curr_model
print(housing_dataframe.columns.values)
newest_data['LOCALITY'] = housing_dataframe['LOCALITY']
newest_model = stats.ols(formula = 'np.log(PRICE) ~ TYPE + B_B + np.log(PROPERTYSQFT) + LOCALITY', data = newest_data).fit()
newest_model.summary()
```

```python
loc_merge = newest_data
loc_merge['LOCALITY'] = loc_merge['LOCALITY'].apply(lambda x: "Queens" if x == "Queens County" else x)
loc_merge['LOCALITY'] = loc_merge['LOCALITY'].apply(lambda x: "New York" if x == "New York County" else x)

```

```python
print(len(loc_merge.loc[loc_merge['LOCALITY'] == "United States"].index.tolist()))
loc_merge.drop(labels = loc_merge.loc[loc_merge['LOCALITY'] == "United States"].index.tolist(), axis = 0, inplace =  True)
loc_merge.drop(labels = loc_merge.loc[loc_merge['LOCALITY'] == "Flatbush"].index.tolist(), axis = 0, inplace =  True)
merge_model = stats.ols(formula = 'np.log(PRICE) ~ TYPE + B_B + np.log(PROPERTYSQFT) + LOCALITY', data = newest_data).fit()
merge_model.summary()
```
