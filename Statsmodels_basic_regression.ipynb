```python
x = df[['interest_rate','unemployment_rate']]
y = df['index_price']
x, y
```

```python
x = sm.add_constant(x)
x
```

```python
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
printed_model = model.summary()
predictions, printed_model
```
